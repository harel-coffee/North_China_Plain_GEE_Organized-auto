##### Define_Basic_Parameters

```python
# define the year_name for test, here choose these years because they represent
# Landsat8/7/5 respectively
year_name     = ['2017_2019','2011_2013', '2008_2010']
percent_value = [0,1,2,3,5,7,10,20,30,50,70,100]
tree_num      = [1] + list(range(10,121,10))
```

```python
combination_name = ['Landsat',
                    'Fourier',
                    'Landsat_Fourier',
                    'Landsat_Mean',
                    'Fourier_Mean',
                    'Landsat_Fourier_Mean']
```

##### Classify Sample_with_img_value

```python
# define the path to point_with_value GEE-Path
path = 'users/Jinzhu_Deakin/North_China_Plain/Sample_with_Landsat_Fourier_Normalized'

# instantiate the combo_instance dictionary
Combo_each_year = {}
Combo_village   = {}

# Create sample_classification instances through [year] --> [Percent] --> [band_combination] --> [Tree]
for year in year_name:

    # merge each year's sample
    Verified_sample  = ee.FeatureCollection(f'{path}/Verified_point_{year}_extract_Landsat_Fourier_Normalized_img')\
                                          .filterMetadata('Built', 'equals', 0)
    Zone_sample      = ee.FeatureCollection(f'{path}/Zone_point_{year}_extract_Landsat_Fourier_Normalized_img')
    Sample_each_year = Verified_sample.merge(Zone_sample)
    
    # import village_sample visually checked at 1990-1992
    Village_point   = ee.FeatureCollection(f'{path}/Village_point_{year}_extract_Landsat_Fourier_Normalized_img')
    Non_built       = ee.FeatureCollection(f'{path}/Invariant_non_built_{year}_extract_Landsat_Fourier_Normalized_img_2017_2019')
    Sample_village  = Village_point.merge(Non_built)
    
    # Get the band_combo names
    Band_combo   = Make_Sample_Classification.Step_2_Create_Band_Combo(Sample_village)

    for combo in Band_combo:
        # for each band-combination, calculate the accuracy using tree_num
        Each_year_instance = Make_Sample_Classification.Step_3_Create_Classification_Instance(year,Sample_each_year,
                                                                                          combo,tree_num,
                                                                                          100,classificaiton_func = Classification )
        Village_instance   = Make_Sample_Classification.Step_3_Create_Classification_Instance(year,Sample_village,
                                                                                          combo,tree_num,
                                                                                          100,classificaiton_func = Classification )
        # add the instance into the dictionary
        Combo_each_year.update(Each_year_instance)
        Combo_village.update(Village_instance)
```

##### Compute and unpack the acc_value 

```python
# innitialize the accuracy dict
Acc_each_year_AND_Village = {}

# innitilize operation flag to report the process
Opts_flat  = 0
Opts_tatal = len(year_name) * len(combination_name) * len(tree_num)

# unpack the combo_instance through [year] --> [band_combination] --> [Percent] --> [Tree]
for year in year_name:
    for name in combination_name:
        for pct in [100]:
            for tree in tree_num:

                Opts_flat+=1

                Acc_each_year_AND_Village[('Each_year',year,name,pct,tree)] = Accuracy_assesment(Combo_each_year[(year,name,pct,tree)])\
                                                                        .Stp_1_Calculate_Accuracy()
                
                Acc_each_year_AND_Village[('Invarient',year,name,pct,tree)]   = Accuracy_assesment(Combo_village[(year,name,pct,tree)])\
                                                                        .Stp_1_Calculate_Accuracy()
                # print out the process
                print(f'{year}_{name}_pct_{pct:03}_tree_{tree:03} accuracy computation finished! ==>{Opts_flat:03}/{Opts_tatal}')
            print('_________________')
```
