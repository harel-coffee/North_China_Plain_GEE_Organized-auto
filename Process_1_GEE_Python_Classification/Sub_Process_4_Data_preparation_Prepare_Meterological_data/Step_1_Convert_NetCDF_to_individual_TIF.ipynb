```python
import arcpy
import os
```

```python
# find all NetCDF files
NetDCF_Folder = './Data_forcing_01yr_010deg'
arcpy.env.workspace = NetDCF_Folder
All_NetCDF = arcpy.ListFiles('*2018*')
```

```python
# list all NetCDF files
All_NetCDF
```

```python
arcpy.env.overwriteOutput = True

for NC in All_NetCDF:
    
    # Copy the NetCDF as multiband tif
    In_NetCDF = NC
    In_NetCDF_name = In_NetCDF.split('_')[0]

    In_NetCDF_Layer = arcpy.MakeNetCDFRasterLayer_md(in_netCDF_file=In_NetCDF, 
                                                     variable=In_NetCDF_name, 
                                                     x_dimension="lon", 
                                                     y_dimension="lat", 
                                                     out_raster_layer="NetCDF_Layer", 
                                                     band_dimension="time", 
                                                     dimension_values="", 
                                                     value_selection_method="BY_VALUE")
    
    # Define the output folder and save NetCDF as a multiband tif there
    Out_NetCDF_path   = os.path.abspath(f'{NetDCF_Folder}/TIFs/{In_NetCDF_name}.tif')
    arcpy.CopyRaster_management(In_NetCDF_Layer, Out_NetCDF_path)

    # Copy each band of the NetCDF generated multi-band tif with year_names
    bandcount = arcpy.GetRasterProperties_management (Out_NetCDF_path, "BANDCOUNT")

    start_year = 1979
    out_path = f'./Data_forcing_01yr_010deg/TIFs'

    for i in range(int(bandcount[0])):

        in_tif = f'{Out_NetCDF_path}\\band_{i + 1}'
        out_name = f'{os.path.abspath(out_path)}\\{In_NetCDF_name}_{start_year + i}.tif'
        
        # save each band to disk
        arcpy.CopyRaster_management(in_tif, out_name)
        
        # print the out_name to indicate the process
        print(out_name)
    
    
    # delete the median files to save disk space
    arcpy.Delete_management(Out_NetCDF_path)
    
```
